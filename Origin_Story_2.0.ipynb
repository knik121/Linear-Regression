```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data= pd.read_csv('Data_Linear_Regression.csv')
```

```python
def cost_function(m,b,points):
    total_error=0
    for i in range(len(points)):
        x=points.iloc[i].High
        y=points.iloc[i].Low
        total_error+= (y-(m*x+b)) ** 2
    total_error/float(len(points))
```

```python
def gradient_descent(mdash,bdash,points,a):
    m_slope=0
    b_slope=0
    n=len(points)
    for i in range(n):
        x=points.iloc[i].High
        y=points.iloc[i].Low
        m_slope= -(1/n)*x*(y-(mdash*x +bdash))
        b_slope= -(1/n)*(y-(mdash*x +bdash))
        
    m=mdash- m_slope*a
    b=bdash- b_slope*a
    
    return m,b
```

```python
m=0
b=0
a=0.001
epochs=10

for i in range(epochs):
    m,b= gradient_descent(m,b,data,a)
```

```python
print(m,b)

plt.scatter(data.High, data.Low, color="red")
plt.plot(list(range(25,350)),[m*x + b for x in range(25,350)], color="black")
plt.show()
```
